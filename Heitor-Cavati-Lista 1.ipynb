```python
#8 - Peça um número de 1 a 7 e informe o dia da semana correspondente.

num = int(input('Digite um número de 1 a 7:'))

if num == 1:
  print('Domingo')
if num == 2:
  print('Segunda-feira')
if num == 3:
  print('Terça-feira')
if num == 4:
  print('Quarta-feira')
if num == 5:
  print('Quinta-feira')
if num == 6:
  print('Sexta-feira')
if num == 7:
  print('Sábado')
```

```python
#10 - Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5).

horas = int(input('Digite a quantidade de horas trabalhadas:'))
valor = float(input('Digite o valor por hora:'))

if horas <= 40:
  salario = horas * valor
  print('O salário é igual a R${salario}')
if horas > 40:
  salario = (horas * valor) + (horas - 40) * (valor * 1.5)
  print('O salário é igual a R${salario}')
```
