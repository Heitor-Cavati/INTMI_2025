```python
#13
mat = [[int(input()) for _ in range(3)] for _ in range(3)]

identidade = True
for i in range(3):
    for j in range(3):
        if (i == j and mat[i][j] != 1) or (i != j and mat[i][j] != 0):
            identidade = False

print("É identidade?" , identidade)
```

```python
#17
m = [[float(input()) for _ in range(3)] for _ in range(3)]

det = (
    m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1])
    - m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0])
    + m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0])
)

print("Determinante =", det)
```

```python
#18
print("Matriz A:")
A = [[int(input()) for _ in range(3)] for _ in range(3)]

print("Matriz B:")
B = [[int(input()) for _ in range(3)] for _ in range(3)]

soma = [[A[i][j] + B[i][j] for j in range(3)] for i in range(3)]
dif  = [[A[i][j] - B[i][j] for j in range(3)] for i in range(3)]

print("Soma:", soma)
print("Diferença:", dif)
```

```python
#29
produtos = {}

for _ in range(3):
    nome = input("Produto: ")
    preços = [float(input("Preço: ")) for _ in range(3)]
    produtos[nome] = preços

maior_var = None
produto_var = ""

for nome, preços in produtos.items():
    var = max(preços) - min(preços)
    if maior_var is None or var > maior_var:
        maior_var = var
        produto_var = nome

print("Maior variação:", produto_var)
```
